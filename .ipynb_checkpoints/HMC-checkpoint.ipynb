```python
import numpy as np
import pandas as pd
import seaborn as sns
import numpy.linalg as la
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def potential_energy(x):
    return x.dot(la.inv(COVARIANCE.T)).dot(x.T)

def grad_potential_energy(x):
    return x.dot(la.inv(COVARIANCE))

def kinetic_energy(p):
    return np.sum(p.T.dot(p)/2)
```

```python
def hamiltonian_monte_carlo(y, leap, delta, N):
    
    for t in range(1,N):

        #   SAMPLE RANDOM MOMENTUM
        p0 = np.random.randn(2).T

        #   SIMULATE HAMILTONIAN DYNAMICS
        #   FIRST 1/2 STEP OF MOMENTUM
        p_star = p0 - 1/2*delta/2*grad_potential_energy(y[t-1,:])

        #   FIRST FULL STEP FOR POSITION/SAMPLE
        y_star = y[t-1,:] + delta*p_star

        #   FULL STEPS
        for i in range(leap):
            # MOMENTUM
            p_star = p_star - delta*grad_potential_energy(y_star)
            # POSITION/SAMPLE
            y_star = y_star + delta*p_star

        p_star = p_star - 1/2*delta*grad_potential_energy(y_star)

        U0 = potential_energy(y[t-1,:])
        U_star = potential_energy(y_star)

        K0 = kinetic_energy(p0)
        K_star = kinetic_energy(p_star)

        alpha = min(1, np.exp((U0 + K0) - (U_star + K_star)))

        y[t,:] = y_star if np.random.random() < alpha else y[t-1,:]
        
    return y
```

```python
# INITIAL STATE for bivariate normal
N = 1000
corr = 0.8
COVARIANCE = np.array([[1,corr],
                       [corr,1]])
x = np.zeros(2*N).reshape(N,2)
x[:] = np.array([4,3])
```

```python
delta = 0.3
leap = 30
x_HMC = hamiltonian_monte_carlo(x.copy(), leap, delta, N)
```
